## Le prélèvement à la source

Il y a quelques jours, Michel Sapin annonçait la mise en oeuvre du prélèvement à la source.

Pour avoir vécu ce mode de prélèvement (en Suisse et au Danemark), j'étais plutôt favorable à cette mesure à cause
de la simplification qu'elle permet. La simplification est telle que j'étais prêt à
oublier que le prélèvement à la source permet de *cacher* l'impôt (et donc de le faire croître sans douleur),
et que c'est un premier pas vers une fusion IR-CSG ubuesque.

Puis j'ai découvert le prélèvement à la source *à la française*.
* Comme on se méfie plus de l'employeur que de la NSA, celui-ci ne connaitra pas le *vrai* taux d'imposition.
* Comme on ne croit plus en la famille et à la solidarité entre époux, l'impôt pourra être individualisé.
* Comme on veut tout changer sans rien changer, on conservera la déclaration d'impôt pour gérer les niches
  fiscales et pour rattraper tous les trous d'une réforme trop vite faite pour l'être bien.
* Pour ceux (de plus en plus nombreux) qui cumulent plusieurs employeurs, je n'ose pas imaginer le choc de
  simplification. 

Bref, la simplification n'est pas au rendez-vous ; les recruteurs de Bercy doivent déjà se frotter les mains.
La simplification ne serait finalement effective que si on renonçait totalement au quotient familial,
ce qui terminerait d'enterrer ce qui faisait la politique familiale de la France.

Tout ce qu'on peut retenir de la réformce c'est :
* Que ça va être compliqué.
* Que ça prépare une fusion IR-CSG qui rendrait la CSG progressive.
* Que ça prépare le coup de grâce porté à la politique familiale.

## Pour pélever à la source, il faut qu'il n'y ait qu'une source.

Pourtant, j'aime l'idée de réformer, de simplifier, de faire bouger les lignes. Alors j'ai imaginé
une autre manière de mettre en oeuvre le prélèvement à la source ; je l'apelle le **Compte Foyer Fiscal (CFF)**.

Le compte foyer fiscal, c'est un compte bancaire qui centralise tous les revenus du foyer fiscal. Il est
alimenté par les employeurs du foyer, qui versent dessus le salaire brut-brut (le *coût total à l'employeur*
que vous voyez sur votre fiche de paye). Le foyer y prélève chaque mois ce dont il a besoin pour vivre, et le
fisc délègue au gestionaire de compte (la banque) la responsabilité d'y prélever les taxes, les impôts et les cotisations. **Tout y est centralisé** : impôt sur le revenu, CSG, charges salariales, charges patronales, cotisation retraite, assurance chomage, ... tout est calculé en fonction des revenus du foyer (ce qui entre
sur le compte), de son train de vie (ce qui sort du compte) et de son quotient familial. Il s'agit donc d'une véritable simplification qui ne remet pas en cause la politique familiale.

En travaillant un peu, on peut imaginer que le CFF soit capable de contenir plusieurs types de placements
plus ou moins avantageux (un peu comme le PEA-PME), et qu'il soit possible de distinguer quel argent appartient
à qui. En effet, si le foyer fiscal contient les revenus d'adolescents qui commencent à travailler, ceux-ci
peuvent légitimement vouloir distinguer leur épargne de celle de leurs parents.

Tout le monde y trouve son compte: l'employeur est libéré du calcul de la feuille de paye, l'état peut fusioner autant d'impôt qu'il le souhaite et ainsi réduire ses coûts, le contribuable est dispensé de la déclaration d'impôt. Noter aussi la disparition de la fastidieuse (et manipulatrice) distinction entre le salaire net, le salaire brut et le salaire brut-brut.

Enfin, c'est une simplification monumentale dans la mesure où **elle permet de rassembler tous les français sous un même régime**. Profession libérale, travailleur *uberisé*, auto-entrepreneur... tous versent leurs revenus sur leur CFF, et tous sont égaux devant le fisc.

## Un impôt sur le train de vie

Un autre effet sympa de la mise en place de ce CFF serait la possibilité de taxer non pas les revenus, mais le train de vie. En effet, le CFF permet au législateur de calculer l'impôt en fonction des revenus du foyer (les rentrées d'argent), mais aussi en fonction de son train de vie (les sorties d'argent). Ainsi, si j'ai de gros revenus
mais que je les laisse sur mon CFF (pour investir dans des entreprises, pour sécuriser ma retraite, pour lisser
mes revenus) je peux être moins taxé que celui qui aurait des revenus identiques et qui les flamberait totalement.

Taxer le train de vie plutôt que le revenu présente plusieurs avantages:
* L'impôt peut être fortement progressif sans être confiscatoire: en effet comme l'épargne et l'investissement
  échappent à l'impôt, les grands revenus (qui sont généralement des investisseurs) peuvent financer l'économie
  sans voir passer le fisc à chaque mouvement de fonds.
* Les revenus du capital peuvent être comptés avec ceux du travail sans forcément être taxés tous les ans.
  Tant qu'ils sont ré-investis dans le CFF, ils échappent à l'impôt ; ils ne seront imposés qu'au moment des
  retraits.
* La progressivité de l'impôt incite les contribuables à conserver un train de vie raisonable et à épargner sur
  le CFF. Cette épargne peut être orientée vers le financement de l'économie, mais elle sert surtout 
  d'auto-assurance sociale: en cas de chute de revenu, je peux piocher dans cette épargne pour conserver mon
  train de vie.
* Le législateur peut facilement créer un impôt négatif qui remplacerait le RSA, la prime d'activité etc. Et qui
  garantirait que le revenu disponible croît quand le revenu primaire croît. Comme tout est centralisé sur le
  CFF, une formule simple permettrait de calculer la valeur de l'impôt et de ne pas laisser croire qu'on pourrait
  gagner plus en travaillant moins.

  Une telle formule pourrait être par exemple:

  $$\textrm{Revenu disponible} = 40 \times \sqrt{\textrm{Revenu primaire}}$$
  
  Cette formule est plus progressive encore que l'IR actuel, mais elle introduit une protection du travailleur
  pauvre qui pourrait permettre (en contrepartie) de supprimer le SMIC dont l'effet de seuil tue l'emploi.
